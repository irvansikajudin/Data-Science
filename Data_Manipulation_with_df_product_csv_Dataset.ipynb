Load Metro_Interstate_Traffic_Volume.csv as DataFrame

From ‘date_time’ column:

*   Create ‘month’ column (10 points)
*   Create ‘day_of_month’ column (digit tanggal) (10 points)
*   Create ‘day_name’ column (Monday, Tuesday, ...) (10 points)
*   Create ‘hour’ column (digit jam) (10 points)
*   Create ‘is_weekend’ binary column (1 if day == saturday OR sunday, 0 otherwise) (10 points) 

```python
df_time = pd.read_csv('/content/gdrive/MyDrive/data science/Metro_Interstate_Traffic_Volume.csv')
df = df_time
df_time.head(5)
```

```python
import numpy as np
import datetime
```

```python
#1 Create ‘month’ column
df['month'] = pd.DatetimeIndex(df['date_time']).month
df[['date_time','month']].head(5)
```

```python
?pd.DatetimeIndex
```

```python
#2 Create ‘day_of_month’ column (digit tanggal)
df['day'] = pd.DatetimeIndex(df['date_time']).day
df[['date_time','day']].head(5)
```

```python
#3 Create ‘day_name’ column (Monday, Tuesday, ...)
df['day_name'] = pd.DatetimeIndex(df['date_time']).day_name()
df[['date_time','day_name']].head(5)
```

```python
#4 Create ‘hour’ column (digit jam) (10 points)
df['hour'] = pd.DatetimeIndex(df['date_time']).hour
df[['date_time','hour']].head(5)
```

```python
#5 Create ‘is_weekend’ binary column (1 if day == saturday OR sunday, 0 otherwise)
df["date_time"] = pd.to_datetime(df["date_time"])
df["DayOfWeek"] = df["date_time"].dt.weekday
df["IsWeekend"] = df["DayOfWeek"] > 5
replacement_mapping_dict = {True:1,False:0}
df["IsWeekend"] = df["IsWeekend"].replace(replacement_mapping_dict)
df[['date_time', 'DayOfWeek','IsWeekend']]
```
